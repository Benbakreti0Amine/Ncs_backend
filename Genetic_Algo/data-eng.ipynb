```python
import pandas as pd

# Your list of standardized commune names
communes = [
    "Ain Benian", "Ain Taya", "Baba Hassen", "Bab El Oued", "Bab Ezzouar",
    "Bachdjerrah", "Baraki", "Ben Aknoun", "Beni Messous", "Birkhadem",
    "Bir Mourad Raïs", "Birtouta", "Bologhine", "Bordj El Bahri",
    "Bordj El Kiffan", "Bourouba", "Bouzereah", "Casbah", "Cheraga",
    "Dar El Beïda", "Dely Ibrahim", "Djasr Kasentina", "Douera", "Draria",
    "El Achour", "El Biar", "El Djazaïr", "El Hammamet", "El Harrach",
    "El Madania", "El Magharia", "El Marsa", "El Mouradia", "Harraoua",
    "Hussein Dey", "Hydra", "Khraicia", "Kouba", "Les Eucalyptus",
    "Mahelma", "Mohamed Belouizdad (Hamma Annassers)", "Mohammadia",
    "Oued Koriche", "Oued Smar", "Ouled Chebel", "Ouled Fayet", "Rahmania",
    "Raïs Hamidou", "Reghaia", "Rouiba", "Saoula", "Sidi M'Hamed",
    "Sidi Moussa", "Souidania", "Staoueli", "Tessala El Merdja", "Zeralda",
    "Algiers"
]

# Assuming df is your messy dataset DataFrame
def match_commune(row):
    for commune in communes:
        if commune.lower().replace("’", "'") in row.lower().replace("’", "'"):
            return commune
    return None

df2['commune_name'] = df2['Country'].apply(match_commune)

df2
```

```python
filtered_df = df2[df2['commune_name'].notna()]
filtered_df=filtered_df[['commune_name','Latitude','Longitude']]
filtered_df
filtered_df = filtered_df[filtered_df["commune_name"] != "Algiers"]
filtered_df

unique_communes = filtered_df.drop_duplicates(subset="commune_name", keep="first")
unique_communes



```

```python
from unidecode import unidecode

# Normalize the join field in both datasets
df['hi'] = df['place_name'].apply(lambda x: unidecode(str(x).lower().strip()))
df
unique_communes['hi']=unique_communes['commune_name'].apply(lambda x: unidecode(str(x).lower().strip()))
unique_communes
```

```python
merged_df = df.merge(unique_communes, on='hi', how='inner')
merged_df
```
